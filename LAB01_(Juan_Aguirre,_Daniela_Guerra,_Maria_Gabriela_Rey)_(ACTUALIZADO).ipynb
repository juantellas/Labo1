```python
import os
from PIL import Image
from graphviz import Digraph

class NodoAVL:
    def __init__(self, nombre_archivo):
        self.nombre_archivo = nombre_archivo
        self.izquierda = None
        self.derecha = None
        self.altura = 1
        self.padre = None

class AVL:
    def __init__(self):
        self.raiz = None


    def obtener_minimo(self, nodo):
      actual = nodo
      while actual.izquierda is not None:
          actual = actual.izquierda
      return actual


    def obtener_altura(self, nodo):
        if not nodo:
            return 0
        return nodo.altura

    def obtener_balance(self, nodo):
        if not nodo:
            return 0
        return self.obtener_altura(nodo.izquierda) - self.obtener_altura(nodo.derecha)

    def rotar_derecha(self, y):
        x = y.izquierda
        T2 = x.derecha

        x.derecha = y
        y.izquierda = T2

        y.altura = 1 + max(self.obtener_altura(y.izquierda), self.obtener_altura(y.derecha))
        x.altura = 1 + max(self.obtener_altura(x.izquierda), self.obtener_altura(x.derecha))

        return x

    def rotar_izquierda(self, x):
        y = x.derecha
        T2 = y.izquierda

        y.izquierda = x
        x.derecha = T2

        x.altura = 1 + max(self.obtener_altura(x.izquierda), self.obtener_altura(x.derecha))
        y.altura = 1 + max(self.obtener_altura(y.izquierda), self.obtener_altura(y.derecha))

        return y

    def splitfilename(self, filename):
      parts = re.split(r'(\d+)', filename)
      numeric_parts = [int(part) if part.isdigit() else part for part in parts if part]
      return numeric_parts


    def insertar(self, raiz, nombre_archivo):
          if not raiz:
              return NodoAVL(nombre_archivo)
          elif nombre_archivo < raiz.nombre_archivo:
              raiz.izquierda = self.insertar(raiz.izquierda, nombre_archivo)
          else:
              raiz.derecha = self.insertar(raiz.derecha, nombre_archivo)

          raiz.altura = 1 + max(self.obtener_altura(raiz.izquierda), self.obtener_altura(raiz.derecha))

          balance = self.obtener_balance(raiz)

          if balance > 1 and nombre_archivo < raiz.izquierda.nombre_archivo:
              return self.rotar_derecha(raiz)

          if balance < -1 and nombre_archivo > raiz.derecha.nombre_archivo:
              return self.rotar_izquierda(raiz)

          if balance > 1 and nombre_archivo > raiz.izquierda.nombre_archivo:
              raiz.izquierda = self.rotar_izquierda(raiz.izquierda)
              return self.rotar_derecha(raiz)

          if balance < -1 and nombre_archivo < raiz.derecha.nombre_archivo:
              raiz.derecha = self.rotar_derecha(raiz.derecha)
              return self.rotar_izquierda(raiz)

          return raiz


    def eliminar(self, raiz, nombre_archivo, padre=None):
      if not raiz:
          return raiz

      if nombre_archivo < raiz.nombre_archivo:
          raiz.izquierda = self.eliminar(raiz.izquierda, nombre_archivo, raiz)
      elif nombre_archivo > raiz.nombre_archivo:
          raiz.derecha = self.eliminar(raiz.derecha, nombre_archivo, raiz)
      else:
          # Nodo con un solo hijo o sin hijos
          if raiz.izquierda is None:
              temp = raiz.derecha
              if padre and padre.izquierda == raiz:
                  padre.izquierda = temp
              else:
                  padre.derecha = temp
              raiz = None
              return temp
          elif raiz.derecha is None:
              temp = raiz.izquierda
              if padre and padre.izquierda == raiz:
                  padre.izquierda = temp
              else:
                  padre.derecha = temp
              raiz = None
              return temp

          # Nodo con dos hijos: obtener el sucesor inorden (el más pequeño en el subárbol derecho)
          temp = self.obtener_minimo(raiz.derecha)

          # Copiar el sucesor inorden al nodo actual
          raiz.nombre_archivo = temp.nombre_archivo

          # Eliminar el sucesor inorden
          raiz.derecha = self.eliminar(raiz.derecha, temp.nombre_archivo, raiz)

      if raiz is None:
          return raiz

      # Actualizar la altura del nodo actual
      raiz.altura = 1 + max(self.obtener_altura(raiz.izquierda), self.obtener_altura(raiz.derecha))

      # Obtener el balance del nodo actual
      balance = self.obtener_balance(raiz)

      # Balancear el árbol si es necesario
      if balance > 1 and self.obtener_balance(raiz.izquierda) >= 0:
          return self.rotar_derecha(raiz)
      if balance < -1 and self.obtener_balance(raiz.derecha) <= 0:
          return self.rotar_izquierda(raiz)
      if balance > 1 and self.obtener_balance(raiz.izquierda) < 0:
          raiz.izquierda = self.rotar_izquierda(raiz.izquierda)
          return self.rotar_derecha(raiz)
      if balance < -1 and self.obtener_balance(raiz.derecha) > 0:
          raiz.derecha = self.rotar_derecha(raiz.derecha)
          return self.rotar_izquierda(raiz)
      return raiz


    def imprimir_inorden(self, nodo):
        if nodo:
            self.imprimir_inorden(nodo.izquierda)
            print(f"Nombre: {nodo.nombre_archivo}")
            self.imprimir_inorden(nodo.derecha)

    def buscar_nodo(self, raiz, nombre_archivo):
      if raiz is None or raiz.nombre_archivo == nombre_archivo:
        return raiz
      if raiz.nombre_archivo < nombre_archivo:
        return self.buscar_nodo(raiz.derecha, nombre_archivo)
      return self.buscar_nodo(raiz.izquierda, nombre_archivo)


    def buscar_nodos_criterios(self, raiz, tipo, peso_min, peso_max, nodos_encontrados=[]):
      if raiz:
          nombre = raiz.nombre_archivo
          categoria = ""
          peso = 0
          for registro in registros_imagenes:
              if registro["Nombre"] == nombre:
                  categoria = registro["Categoria"]
                  peso = registro["Peso"]
                  break

          if categoria == tipo and peso >= peso_min and peso < peso_max:
              nodos_encontrados.append({"Nombre": nombre, "Categoria": categoria, "Peso": peso})

          self.buscar_nodos_criterios(raiz.izquierda, tipo, peso_min, peso_max, nodos_encontrados)
          self.buscar_nodos_criterios(raiz.derecha, tipo, peso_min, peso_max, nodos_encontrados)

      return nodos_encontrados

    def graficar_arbol(self, nodo, dot, registros_imagenes):
        if nodo:
            nombre = nodo.nombre_archivo
            categoria = ""
            peso = ""
            for registro in registros_imagenes:
                if registro["Nombre"] == nombre:
                    categoria = registro["Categoria"]
                    peso = registro["Peso"]
                    break

            label = f"Nombre: {nombre}\nCategoria: {categoria}\nPeso: {peso:.2f} MB"
            dot.node(nombre, label=label)
            if nodo.izquierda:
                dot.edge(nombre, nodo.izquierda.nombre_archivo)
                self.graficar_arbol(nodo.izquierda, dot, registros_imagenes)
            if nodo.derecha:
                dot.edge(nombre, nodo.derecha.nombre_archivo)
                self.graficar_arbol(nodo.derecha, dot, registros_imagenes)


    def vaciar_arbol(self):
        self.raiz = None

    def obtener_padre(self, nombre_archivo):
        nodo = self.obtener_nodo(self.raiz, nombre_archivo)
        if nodo and nodo.padre:
            return nodo.padre.nombre_archivo
        return None


    def obtener_abuelo(self, nombre_archivo):
        nodo = self.obtener_nodo(self.raiz, nombre_archivo)
        if nodo and nodo.padre and nodo.padre.padre:
            return nodo.padre.padre.nombre_archivo
        return None


    def obtener_tio(self, nombre_archivo):
        nodo = self.obtener_nodo(self.raiz, nombre_archivo)
        if nodo and nodo.padre and nodo.padre.padre:
            padre = nodo.padre
            abuelo = padre.padre
            if abuelo.izquierda == padre:
                tio = abuelo.derecha
            else:
                tio = abuelo.izquierda
            if tio:
                return tio.nombre_archivo
        return None


    def obtener_padre(self, nodo):
                    print("a")
                    if nodo:
                       return self._obtener_padre_recur(self.raiz, nodo, None)
                    else:
                      return None

    def _obtener_padre_recur(self, raiz, nodo, padre_actual):
                        if raiz is None:
                            return None
                        elif raiz.izq == nodo or raiz.der == nodo:
                            return raiz
                        elif nodo.nombre < raiz.nombre:
                            return self._obtener_padre_recur(raiz.izq, nodo, raiz)
                        else:
                            return self._obtener_padre_recur(raiz.der, nodo, raiz)


    def buscar_nodos_criterios(self, raiz, tipo, peso_min, peso_max, nodos_encontrados=None):
        if nodos_encontrados is None:
            nodos_encontrados = []

        if raiz is None:
            return nodos_encontrados

        if tipo == raiz.categoria and peso_min <= raiz.peso <= peso_max:
            nodos_encontrados.append(raiz)

        if raiz.left:
            self.buscar_nodos_criterios(raiz.left, tipo, peso_min, peso_max, nodos_encontrados)

        if raiz.right:
            self.buscar_nodos_criterios(raiz.right, tipo, peso_min, peso_max, nodos_encontrados)

        return nodos_encontrados

    def obtener_padre(nodo, nombre_archivo):
                  if nodo is None:
                      return None
                  if nodo.nombre_archivo == nombre_archivo:
                      return nodo.padre.nombre_archivo if nodo.padre else None
                  izq = obtener_padre(nodo.izquierda, nombre_archivo)
                  if izq is not None:
                      return izq
                  der = obtener_padre(nodo.derecha, nombre_archivo)
                  return der

```
