##Creacion del Arbol AVL##

```python
def separador():
  print("----------------------------------------")

class NodoAVL:
    def __init__(self, nombre_archivo):
        self.nombre_archivo = nombre_archivo
        self.izquierda = None
        self.derecha = None
        self.altura = 1
        self.padre = None

class AVL:
    def __init__(self):
        self.raiz = None

    def obtener_minimo(self, nodo):
      actual = nodo
      while actual.izquierda is not None:
          actual = actual.izquierda
      return actual


#Funcion encargada de obtener la altura de un nodo
    def obtener_altura(self, nodo):
        if not nodo:
            return 0
        return nodo.altura



#Funcion encargada de obtener el nivel de balanceo de un nodo
    def obtener_balance(self, nodo):
        if not nodo:
            return 0
        return self.obtener_altura(nodo.derecha) - self.obtener_altura(nodo.izquierda)



#Funcion encargada de rotar los nodos del arbol AVL a la derecha
    def rotar_derecha(self, y):
        x = y.izquierda
        T2 = x.derecha

        x.derecha = y
        y.izquierda = T2

        y.altura = 1 + max(self.obtener_altura(y.izquierda), self.obtener_altura(y.derecha))
        x.altura = 1 + max(self.obtener_altura(x.izquierda), self.obtener_altura(x.derecha))

        return x

#Funcion encargada de rotar los nodos del arbol AVL a la izquierda
    def rotar_izquierda(self, x):
        y = x.derecha
        T2 = y.izquierda

        y.izquierda = x
        x.derecha = T2

        x.altura = 1 + max(self.obtener_altura(x.izquierda), self.obtener_altura(x.derecha))
        y.altura = 1 + max(self.obtener_altura(y.izquierda), self.obtener_altura(y.derecha))

        return y


#Funcion encargada de separar las partes numericas de las que no son (para comparar y ordenar de manera alfanumerica)
    def splitfilename(self, filename):
        parts = re.split(r'(\d+)', filename)
        normalized_parts = []
        for part in parts:
            if part.isdigit():
                normalized_parts.append(int(part))
            else:
                normalized_parts.append(part.lower().replace("-", "").replace("_", ""))
        return normalized_parts


#Funcion encargada para insertar nodos dentro del arbol
    def insertar(self, raiz, nombre_archivo, padre=None):

      if not raiz:
          nuevo_nodo = NodoAVL(nombre_archivo)
          nuevo_nodo.padre = padre
          return NodoAVL(nombre_archivo)

      nombre_archivo_parts = self.splitfilename(nombre_archivo)
      raiz_parts = self.splitfilename(raiz.nombre_archivo)

      if nombre_archivo_parts < raiz_parts:
          raiz.izquierda = self.insertar(raiz.izquierda, nombre_archivo)
          raiz.izquierda.padre = raiz
      else:
          raiz.derecha = self.insertar(raiz.derecha, nombre_archivo)
          raiz.derecha.padre = raiz

      raiz.altura = 1 + max(self.obtener_altura(raiz.izquierda), self.obtener_altura(raiz.derecha))

      balance = self.obtener_balance(raiz)

      if balance > 1 and nombre_archivo_parts < self.splitfilename(raiz.izquierda.nombre_archivo):
          return self.rotar_derecha(raiz)

      if balance < -1 and nombre_archivo_parts > self.splitfilename(raiz.derecha.nombre_archivo):
          return self.rotar_izquierda(raiz)

      if balance > 1 and nombre_archivo_parts > self.splitfilename(raiz.izquierda.nombre_archivo):
          raiz.izquierda = self.rotar_izquierda(raiz.izquierda)
          return self.rotar_derecha(raiz)

      if balance < -1 and nombre_archivo_parts < self.splitfilename(raiz.derecha.nombre_archivo):
          raiz.derecha = self.rotar_derecha(raiz.derecha)
          return self.rotar_izquierda(raiz)

      return raiz


#Funcion encargada de eliminar nodos dentro del arbol AVL
    def eliminar(self, raiz, nombre_archivo, padre=None):
      if not raiz:
          return raiz

      if nombre_archivo < raiz.nombre_archivo:
          raiz.izquierda = self.eliminar(raiz.izquierda, nombre_archivo, raiz)
      elif nombre_archivo > raiz.nombre_archivo:
          raiz.derecha = self.eliminar(raiz.derecha, nombre_archivo, raiz)
      else:
          # Nodo con un solo hijo o sin hijos
          if raiz.izquierda is None:
              temp = raiz.derecha
              if padre and padre.izquierda == raiz:
                  padre.izquierda = temp
              else:
                  padre.derecha = temp
              raiz = None
              return temp
          elif raiz.derecha is None:
              temp = raiz.izquierda
              if padre and padre.izquierda == raiz:
                  padre.izquierda = temp
              else:
                  padre.derecha = temp
              raiz = None
              return temp

          temp = self.obtener_minimo(raiz.derecha)

          raiz.nombre_archivo = temp.nombre_archivo

          raiz.derecha = self.eliminar(raiz.derecha, temp.nombre_archivo, raiz)

      if raiz is None:
          return raiz

      raiz.altura = 1 + max(self.obtener_altura(raiz.izquierda), self.obtener_altura(raiz.derecha))

      balance = self.obtener_balance(raiz)

      if balance > 1 and self.obtener_balance(raiz.izquierda) >= 0:
          return self.rotar_derecha(raiz)
      if balance < -1 and self.obtener_balance(raiz.derecha) <= 0:
          return self.rotar_izquierda(raiz)
      if balance > 1 and self.obtener_balance(raiz.izquierda) < 0:
          raiz.izquierda = self.rotar_izquierda(raiz.izquierda)
          return self.rotar_derecha(raiz)
      if balance < -1 and self.obtener_balance(raiz.derecha) > 0:
          raiz.derecha = self.rotar_derecha(raiz.derecha)
          return self.rotar_izquierda(raiz)
      return raiz

#Funcion encargada de buscar un nodo con base a su nombre.
    def buscar_nodo(self, raiz, nombre_archivo):
      if raiz is None or raiz.nombre_archivo == nombre_archivo:
        return raiz
      if raiz.nombre_archivo < nombre_archivo:
        return self.buscar_nodo(raiz.derecha, nombre_archivo)
      return self.buscar_nodo(raiz.izquierda, nombre_archivo)


#Funcion encargada de buscar los nodos con base a un criterio (categoria y peso)
    def buscar_nodos_criterios(self, raiz, tipo, peso_min, peso_max, nodos_encontrados=[]):
      if raiz:
          nombre = raiz.nombre_archivo
          categoria = ""
          peso = 0
          for registro in registros_imagenes:
              if registro["Nombre"] == nombre:
                  categoria = registro["Categoria"]
                  peso = registro["Peso"]
                  break

          if categoria == tipo and peso >= peso_min and peso < peso_max:
              nodos_encontrados.append({"Nombre": nombre, "Categoria": categoria, "Peso": peso})

          self.buscar_nodos_criterios(raiz.izquierda, tipo, peso_min, peso_max, nodos_encontrados)
          self.buscar_nodos_criterios(raiz.derecha, tipo, peso_min, peso_max, nodos_encontrados)

      return nodos_encontrados


#Funcion encargada de graficar, usando la libreria graphviz, el arbol AVL
    def graficar_arbol(self, nodo, dot, registros_imagenes):
        if nodo:
            nombre = nodo.nombre_archivo
            categoria = ""
            peso = ""
            for registro in registros_imagenes:
                if registro["Nombre"] == nombre:
                    categoria = registro["Categoria"]
                    peso = registro["Peso"]
                    break

            label = f"Nombre: {nombre}\nCategoria: {categoria}\nPeso: {peso:.2f} MB"
            dot.node(nombre, label=label)
            if nodo.izquierda:
                dot.edge(nombre, nodo.izquierda.nombre_archivo)
                self.graficar_arbol(nodo.izquierda, dot, registros_imagenes)
            if nodo.derecha:
                dot.edge(nombre, nodo.derecha.nombre_archivo)
                self.graficar_arbol(nodo.derecha, dot, registros_imagenes)


#Funcion encargada de vaciar el arbol AVL
    def vaciar_arbol(self):
        self.raiz = None




```

```python
print("Bienvenido al algoritmo del Arbol Binario Balanceado.")
separador()
print("Creadores: Juan Aguirre, Daniela Guerra, Maria Gabriela Rey.")
separador()

user=str(input("Por favor, ingrese su nombre:\n"))

option = 0

#Opciones del Menu
while option != 6:
    separador()
    print("Hola", user, ", Que opcion te gustaria utilizar hoy?")
    separador()
    print("#1. Aadir Nodo. ")
    print("#2. Eliminar Nodo. ")
    print("#3. Buscar Nodo.")
    print("#4. Buscar Nodo (con parametros especificos).")
    print("#5. Mostrar el recorrido por niveles del arbol.")
    print("#6. Salir del programa.")
    separador()

    option = int(input("Seleccione una opcin: "))
    separador()

#-----------------------------------------------------------------------------------------------------

    if option == 1:
        print("Eleccion escogida: [#1. Insertar Nodo.]")
        nombre_archivo = input("Ingrese el nombre del nodo a insertar: ")

        #Sequencia que revisa si los nodos existen y estan en la carpeta con las imagenes/registros
        if os.path.exists(f"/content/dogs/{nombre_archivo}"):
            avl.raiz = avl.insertar(avl.raiz, nombre_archivo)
            print("El nodo ha sido insertado exitosamente.")
            separador()
            dot = Digraph(comment="rbol AVL de nombres de archivos")
            avl.graficar_arbol(avl.raiz, dot, registros_imagenes)
            dot.render("arbol_avl_nombres", format="png", cleanup=True)
            dot.view()
        else:
            print(f"El archivo '{nombre_archivo}' no existe en el directorio '/content/dogs'.")
            separador()

#-----------------------------------------------------------------------------------------------------

    elif option == 2:
        print("Eleccion escogida: [#2. Eliminar Nodo.]")
        nombre_archivo = input("Ingrese el nombre del nodo a eliminar: ")

        #Sequencia que revisa si los nodos existen y estan en la carpeta con las imagenes/registros
        if os.path.exists(f"/content/dogs/{nombre_archivo}"):
            avl.raiz = avl.eliminar(avl.raiz, nombre_archivo, None)
            print("El nodo ha sido eliminado exitosamente.")
            separador()
            dot = Digraph(comment="rbol AVL de nombres de archivos")
            avl.graficar_arbol(avl.raiz, dot, registros_imagenes)
            dot.render("arbol_avl_nombres", format="png", cleanup=True)
            dot.view()
        else:
            print(f"El archivo '{nombre_archivo}' no existe en el directorio '/content/dogs'.")
            separador()

#-----------------------------------------------------------------------------------------------------

    elif option == 3:
      print("Eleccion escogida: [#3. Buscar Nodo..]")
      nombre_archivo = input("Ingrese el nombre del nodo a buscar: ")
      nodo_encontrado = avl.buscar_nodo(avl.raiz, nombre_archivo)

      if nodo_encontrado:
          print(f"El nodo con nombre de archivo '{nombre_archivo}' ha sido encontrado.")
          separador()
          option2=0

          while option2!=-1:
            print("Que accion le gustaria hacer con:", nombre_archivo, "?")
            separador()
            print("#1. Buscar nivel del nodo. ")
            print("#2. Averiguar el factor de balanceo del nodo. ")
            print("#3. Buscar el padre del nodo.")
            print("#4. Buscar el abuelo del nodo.")
            print("#5. Buscar el tio del nodo.")
            print("#6. Salir de estas opciones.")
            separador()

            option2=int(input("Por favor, elija una de las opciones disponibles \n"))
            separador()

            #-----------------------------------------------------------------------------------------------------

            if option2 == 1:
                print("Eleccion escogida: [#1. Obtener nivel del nodo.]")
                separador()

                def buscar_nivel(nodo, nombre_archivo, nivel_actual=1):
                    if nodo is None:
                        return None
                    if nodo.nombre_archivo == nombre_archivo:
                        return nivel_actual
                    izq = buscar_nivel(nodo.izquierda, nombre_archivo, nivel_actual + 1)
                    if izq is not None:
                        return izq
                    der = buscar_nivel(nodo.derecha, nombre_archivo, nivel_actual + 1)
                    if der is not None:
                        return der-1
                    return None

                nivel = buscar_nivel(avl.raiz, nombre_archivo)
                if nivel is not None:
                    print(f"El nivel del nodo con nombre de archivo '{nombre_archivo}' es: {nivel}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()

            #-----------------------------------------------------------------------------------------------------

            elif option2 == 2:
                print("Eleccion escogida: [#2. Obtener factor de balance del nodo.]")
                nodo = avl.buscar_nodo(avl.raiz, nombre_archivo)
                separador()

                if nodo:
                    altura_izquierda = avl.obtener_altura(nodo.izquierda)
                    altura_derecha = avl.obtener_altura(nodo.derecha)
                    factor_balance = altura_derecha - altura_izquierda
                    print(f"El factor de balance del nodo '{nombre_archivo}' es: {factor_balance}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()

            #-----------------------------------------------------------------------------------------------------

            elif option2 == 3:
              print("Eleccion escogida: [#3. Obtener padre del nodo.]")

              def obtener_padre(nodo, nombre_archivo):
                  if nodo is None:
                      return None
                  if nodo.nombre_archivo == nombre_archivo:
                      return nodo.padre.nombre_archivo if nodo.padre else None
                  izq = obtener_padre(nodo.izquierda, nombre_archivo)
                  if izq is not None:
                      return izq
                  der = obtener_padre(nodo.derecha, nombre_archivo)
                  return der

              padre = obtener_padre(avl.raiz, nombre_archivo)
              if padre is not None:
                  print(f"El padre del nodo con nombre de archivo '{nombre_archivo}' es: {padre}")
              else:
                  print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
              separador()

            #-----------------------------------------------------------------------------------------------------

            elif option2 == 4:
                print("Eleccion escogida: [#4. Obtener abuelo del nodo.]")
                separador()

                def obtener_abuelo(nodo, nombre_archivo):
                    if nodo is None or nodo.nombre_archivo == nombre_archivo:
                        return nodo.padre.padre.nombre_archivo if nodo and nodo.padre and nodo.padre.padre else None
                    izq = obtener_abuelo(nodo.izquierda, nombre_archivo)
                    if izq is not None:
                        return izq
                    der = obtener_abuelo(nodo.derecha, nombre_archivo)
                    if der is not None:
                        return der
                    return None

                abuelo = obtener_abuelo(avl.raiz, nombre_archivo)
                if abuelo is not None:
                    print(f"El abuelo del nodo con nombre de archivo '{nombre_archivo}' es: {abuelo}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()
            #-----------------------------------------------------------------------------------------------------

            elif option2 == 5:
              print("Eleccion escogida: [#5. Obtener tio del nodo.]")
              separador()

              def obtener_tio(nodo, nombre_archivo):
                  if nodo is None or nodo.nombre_archivo == nombre_archivo:
                      return None

                  padre = nodo.padre
                  if padre is None or padre.nombre_archivo == nombre_archivo:
                      return None

                  abuelo = padre.padre
                  if abuelo is None:
                      return None

                  tio = abuelo.izquierda if abuelo.izquierda != padre else abuelo.derecha if abuelo.derecha else None
                  return tio.nombre_archivo if tio else None

              tio = obtener_tio(avl.raiz, nombre_archivo)
              if tio is not None:
                  print(f"El tio del nodo con nombre de archivo '{nombre_archivo}' es: {tio}")
                  separador()
              else:
                  print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                  separador()

            #-----------------------------------------------------------------------------------------------------

            elif option2 == 6:
              print("Eleccion escogida: [#6. Salir de estas opciones.]")
              break

          else:
            print("Por favor, escoja una opcion valida")

            #-----------------------------------------------------------------------------------------------------
      else:
          print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
      separador()

#-----------------------------------------------------------------------------------------------------

    elif option == 4:
      print("Eleccion escogida: [#4. Buscar Nodo (con parametros especificos).]")
      tipo = input("Ingrese el tipo (dogs, humans, horses, cats, bikes, cars, flower): ")
      peso_min = float(input("Ingrese el peso mnimo en MB: "))
      peso_max = float(input("Ingrese el peso mximo en MB: "))
      separador()

      nodos_encontrados = avl.buscar_nodos_criterios(avl.raiz, tipo, peso_min, peso_max)

      print("Los nodos encontrados fueron los siguientes:")
      for nodo in nodos_encontrados:
          print(f"Nombre: {nodo['Nombre']}, Categoria: {nodo['Categoria']}, Peso: {nodo['Peso']:.2f} MB")
      separador()

      nombre_archivo = input("Ingrese el nombre del nodo que desea usar: ")
      separador()
      option3=0

      while option3!=-1:
            print("Que accion le gustaria hacer con:", nombre_archivo, "?")
            separador()
            print("#1. Buscar nivel del nodo. ")
            print("#2. Averiguar el factor de balanceo del nodo. ")
            print("#3. Buscar el padre del nodo.")
            print("#4. Buscar el abuelo del nodo.")
            print("#5. Buscar el tio del nodo.")
            print("#6. Salir de estas opciones.")
            separador()

            option3=int(input("Por favor, elija una de las opciones disponibles \n"))
            separador()

            if option3 == 1:
                print("Eleccion escogida: [#1. Obtener nivel del nodo.]")
                separador()

                def buscar_nivel(nodo, nombre_archivo, nivel_actual=1):
                    if nodo is None:
                        return None
                    if nodo.nombre_archivo == nombre_archivo:
                        return nivel_actual
                    izq = buscar_nivel(nodo.izquierda, nombre_archivo, nivel_actual + 1)
                    if izq is not None:
                        return izq
                    der = buscar_nivel(nodo.derecha, nombre_archivo, nivel_actual + 1)
                    if der is not None:
                        return der-1
                    return None

                nivel = buscar_nivel(avl.raiz, nombre_archivo)
                if nivel is not None:
                    print(f"El nivel del nodo con nombre de archivo '{nombre_archivo}' es: {nivel}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()

            #-----------------------------------------------------------------------------------------------------

            elif option3 == 2:
                print("Eleccion escogida: [#2. Obtener factor de balance del nodo.]")
                nodo = avl.buscar_nodo(avl.raiz, nombre_archivo)
                separador()

                if nodo:
                    altura_izquierda = avl.obtener_altura(nodo.izquierda)
                    altura_derecha = avl.obtener_altura(nodo.derecha)
                    factor_balance = altura_derecha - altura_izquierda
                    print(f"El factor de balance del nodo '{nombre_archivo}' es: {factor_balance}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()

            #-----------------------------------------------------------------------------------------------------

            elif option3 == 3:
              print("Eleccion escogida: [#3. Obtener padre del nodo.]")

              def obtener_padre(nodo, nombre_archivo):
                  if nodo is None:
                      return None
                  if nodo.nombre_archivo == nombre_archivo:
                      return nodo.padre.nombre_archivo if nodo.padre else None
                  izq = obtener_padre(nodo.izquierda, nombre_archivo)
                  if izq is not None:
                      return izq
                  der = obtener_padre(nodo.derecha, nombre_archivo)
                  return der

              padre = obtener_padre(avl.raiz, nombre_archivo)
              if padre is not None:
                  print(f"El padre del nodo con nombre de archivo '{nombre_archivo}' es: {padre}")
              else:
                  print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
              separador()

            #-----------------------------------------------------------------------------------------------------

            elif option3 == 4:
                print("Eleccion escogida: [#4. Obtener abuelo del nodo.]")
                separador()

                def obtener_abuelo(nodo, nombre_archivo):
                    if nodo is None or nodo.nombre_archivo == nombre_archivo:
                        return nodo.padre.padre.nombre_archivo if nodo and nodo.padre and nodo.padre.padre else None
                    izq = obtener_abuelo(nodo.izquierda, nombre_archivo)
                    if izq is not None:
                        return izq
                    der = obtener_abuelo(nodo.derecha, nombre_archivo)
                    if der is not None:
                        return der
                    return None

                abuelo = obtener_abuelo(avl.raiz, nombre_archivo)
                if abuelo is not None:
                    print(f"El abuelo del nodo con nombre de archivo '{nombre_archivo}' es: {abuelo}")
                    separador()
                else:
                    print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                    separador()

            #-----------------------------------------------------------------------------------------------------

            elif option3 == 5:
              print("Eleccion escogida: [#5. Obtener tio del nodo.]")
              separador()

              def obtener_tio(nodo, nombre_archivo):
                  if nodo is None or nodo.nombre_archivo == nombre_archivo:
                      return None

                  padre = nodo.padre
                  if padre is None or padre.nombre_archivo == nombre_archivo:
                      return None

                  abuelo = padre.padre
                  if abuelo is None:
                      return None

                  tio = abuelo.izquierda if abuelo.izquierda != padre else abuelo.derecha if abuelo.derecha else None
                  return tio.nombre_archivo if tio else None

              tio = obtener_tio(avl.raiz, nombre_archivo)
              if tio is not None:
                  print(f"El tio del nodo con nombre de archivo '{nombre_archivo}' es: {tio}")
                  separador()
              else:
                  print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
                  separador()

            #-----------------------------------------------------------------------------------------------------

            elif option3 == 6:
              print("Eleccion escogida: [#6. Salir de estas opciones.]")
              break

            else:
              print("Por favor, escoja una opcion valida")

            #-----------------------------------------------------------------------------------------------------

      else:
          print(f"No se encontr el nodo con nombre de archivo '{nombre_archivo}'.")
      separador()

#-----------------------------------------------------------------------------------------------------

    elif option == 5:
      print("Eleccion escogida: [#5. Mostrar el recorrido por niveles del arbol.]")
      separador()

      def recorrer_niveles(nodo):
          if nodo is None:
              return

          cola = deque()
          cola.append(nodo)

          while cola:
              nodo_actual = cola.popleft()
              print(nodo_actual.nombre_archivo)

              if nodo_actual.izquierda:
                  cola.append(nodo_actual.izquierda)
              if nodo_actual.derecha:
                  cola.append(nodo_actual.derecha)

      recorrer_niveles(avl.raiz)
      separador()

#-----------------------------------------------------------------------------------------------------

    elif option == 6:
        print("Eleccion escogida: [#6. Salir del programa y vaciar el rbol.]")
        avl.vaciar_arbol()
        print("El rbol ha sido vaciado.")
        print("Hasta luego!")
        separador()

    else:
        print("Opcin invlida. Por favor, seleccione una opcin vlida.")
```
